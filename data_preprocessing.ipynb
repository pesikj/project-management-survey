```python
import pandas as pd


data = pd.read_excel("survey-data-focus-group-digipm (1).xlsx")
data = data.fillna(0)
data.head()
```

```python
data = data.loc[:, "Potvrdit, prioritizovat a revidovat kritéria úspěchu projektu. x není potřeba v realitě x je potřeba v realitě":"Posoudit neformální kulturu a hodnoty podniku a jejich důsledky pro projekt. x nepatří do výuky PM x patří do výuky PM"]
```

```python
data_mean_sorted = data.mean().sort_values(ascending=False)
```

```python
tab_1 = data_mean_sorted[data_mean_sorted.index.str.contains("je potřeba v realitě")]
tab_1.index = tab_1.index.str.split(' x ', n=1).str[0].str.strip()
tab_1 = tab_1.to_frame("potřeba")
tab_1
```

```python
tab_2 = data_mean_sorted[data_mean_sorted.index.str.contains("patří do výuky PM")]
tab_2.index = tab_2.index.str.split('x', n=1).str[0].str.strip()
tab_2 = tab_2.to_frame("patří")
tab_2
```

```python
merged = pd.merge(tab_1, tab_2, left_index=True, right_index=True)
merged = merged.reset_index()
merged
```

```python
protocol = pd.read_excel("focus group protokol.xlsx")
merged_2 = pd.merge(merged, protocol, left_on="index", right_on="CZ verze", how="left")
merged_2.shape
merged_2.head()
```

```python
merged_2 = merged_2[["index", "potřeba", "patří", "Practice"]]
merged_2.to_excel("merged_2.xlsx")
merged_2
```

```python
grouped_by_practice = merged_2.groupby("Practice")[["potřeba", "patří"]].mean()
grouped_by_practice.sort_values("potřeba", ascending=False)
grouped_by_practice.to_excel("practice.xlsx")
```
